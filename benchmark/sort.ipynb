```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sort.csv')
print(df)
```

```python

entropies = list(set(df['entropy']))
for entropy in entropies:
    data = df[df['entropy']==entropy].groupby(['program','size'])
    means = data['time (ns)'].mean().unstack().transpose()
    ax = means.plot(title='Time to sort with dupplicate-removing \n random pairs arrays with {}% of dupplicatas.'.format(entropy*100),loglog=True)
    ax
```
