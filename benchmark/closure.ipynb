```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('closure.csv')
df
```

```python
tools = ('inferray', 'inferrust')
plt.figure(figsize=(12,6))
for tool in tools:
    plt.plot(df[df['program']==tool]['size'], df[df['program']==tool]['time'], label=tool)
plt.yscale('log')
plt.xscale('log')
plt.ylabel('time')
plt.xlabel('size of the chain of transitive properties')
plt.title('Comparison of transitive-closure of RDF Graph')
plt.legend()
plt.savefig('closure.png')
```
