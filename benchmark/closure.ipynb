```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('closure.csv')
df
```

```python
data = df.groupby(['program','size'])
means = data['time'].mean().unstack().transpose()
stdev = data['time'].std().unstack().transpose()
ax = means.plot(yerr=2*stdev,title='Comparison of transitive-closure of RDF gaph',loglog=True,figsize=(12,6))
ax.set_ylabel('time (in s)')
ax.set_xlabel('length of the transitive-properties chain')
ax.get_figure().savefig("closure.png")
ax
```
